**ПРОЕКТ: Разработка решения для маркетплейса "В один клик", позволяющего кастомизировать предложения для постоянных клиентов, с целью повышения их покупательской активности**

Описание проекта: 

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.

Цели проекта: 

+ Изучить представленные данные и подготовить для обучения на них моделей.
+ Разработать модель, предсказывающую вероятность снижения покупательской активности клиента в следующие три месяца.
+ С помощью обученных моделей и дополнительных данных о прибыльности клиентов, сегментироавть покупателей и разработать для них персонализированные предложения, чтобы увеличить их покупательскую активность.

План работы:

- Анализ важности признаков

## **Шаг 6. Использование пайплайнов**

+ Требуется создать пайплайн для отбора лучшей модели с подходящими под цель исследования гиперпараметрами.

Модели заданы по условиям задачи, гиперпараметры подберем самостоятельно:

**DecisionTreeClassifier()** с гиперпараметрами:
max_depth (максимальная глубина дерева решений) от 2 до 5 (во избежание переобучения модели)
max_features (размер случайных подмножеств объектов, которые необходимо учитывать при разделении узла) от 2 до 5 

**KNeighborsClassifier()** с гиперпараметром:
n_neighbors (количество ближайших соседей) от 2 до 5 

**LogisticRegression()**:
с решателем solver='liblinear' (Он подходит для небольших и средних наборов данных и отличается эффективностью и точностью. Особенно эффективен при решении задач двоичной классификации) и лассо-регуляризацией с гиперпараметром:
С от 1 до 5 

**SVC()**:
с линейным ядром kernel='linear' (простейшее ядро, которое строит гиперплоскость для разделения данных)
с гиперпараметром: перебор «силы» модели для ядра: от 1 до 5 

После выбора лучшей модели произведем расчет **ROC-AUC** на тренировочной и тестовой выборках. Эта метрика оптимально
подходит для бинарной классификации при дисбалансе в классах. 

```python
markt_comm.info()
```

```python
#разделим признаки по методу обработки
#кодирование с OneHotEncoder (категориальные признаки)
ohe_columns = ['тип_сервиса', 'разрешить_сообщать', 'популярная_категория']
#масштабирование/стандартизация (количественные признаки)
num_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность',
               'акционные_покупки', 'категорий_за_визит',
               'неоплаченные_товары', 'ошибка_сервиса', 'страниц_за_визит', 
               'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц', 
               'минут_предыдущий_месяц', 'минут_текущий_месяц']
```

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25

X = markt_comm.drop(columns=['покупательская_активность', 'id'], axis=1) 
y = markt_comm['покупательская_активность'] 

X_train, X_test, y_train, y_test =train_test_split(
    X, 
    y, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE,
    stratify = markt_comm['покупательская_активность']
)
#категориальные признаки
ohe_columns = [
    'тип_сервиса', 'разрешить_сообщать', 'популярная_категория'
]

#количественные признаки
num_columns = [
    'маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность',
               'акционные_покупки', 'категорий_за_визит',
               'неоплаченные_товары', 'ошибка_сервиса', 'страниц_за_визит', 
               'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц', 
               'минут_предыдущий_месяц', 'минут_текущий_месяц'
]

#добаляем ohe-кодирование
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
)
#пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
#итоговый пайплайн для обучения
pipe_final= Pipeline( 
    [
 ('preprocessor', data_preprocessor), 
 ('models', DecisionTreeClassifier(random_state=RANDOM_STATE)) 
    ]
)
#словари для обязательных моделей
param_grid = [
        {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
         {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='linear', probability=True)],
        'models__C': range(1, 5),        
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    }      
]
#определение лучшей модели и параметров
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    n_iter=5,
    random_state=RANDOM_STATE,
    n_jobs=-1

)
# обучение лучшей модели
randomized_search.fit(X_train, y_train)     

#лучшая модель и значение ROC-AUC на тренировочной выборке
print('Лучшая модель:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', round(randomized_search.best_score_, 2))
```

```python
#проверка модели на тестовой выборке
y_test_pred = randomized_search.predict(X_test)
y_test_proba = randomized_search.predict_proba(X_test)
print(f'Значение ROC-AUC на тестовой выборке: {round(roc_auc_score(y_test, y_test_proba[:,1]), 2)}')
```

```python
#итоги по среднему тестовому значению
pd.DataFrame(randomized_search.cv_results_)[
    ['rank_test_score', 'mean_test_score', 'param_models', 'params','std_test_score']
].sort_values('rank_test_score')
```

**Вывод**

Таким образом, найдена и обучена лучшая модель для изучения динамики покупательской активности на сайте. Лучшие результаты показала модель **LogisticRegression(C=1, penalty='l1', random_state=42, solver='liblinear')**. 

Метрика **roc_auc** для тренировочной выборки показла результат 0.90, для тестовой выборки 0.92. Это высокие показатели, соответственно качество предсказаний модели вызывает доверие. В пользу данного вывода говорит как рост значения метрики **roc_auc** на тестовой выборке, по сравнению с тренировочной, так и показатель метрики **f1**, равный 0.90.
